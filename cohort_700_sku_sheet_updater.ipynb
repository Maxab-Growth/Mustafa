# Cohort 700 SKU Data Google Sheet Updater

This notebook:
1. Fetches SKU data for cohort 700 from Snowflake
2. Adds a **date column** with the following logic:
   - If SKU exists in both sheet and query → preserves the existing date from sheet
   - If SKU is new (only in query) → uses today's date
   - If no date in sheet → uses today's date
3. Updates a Google Sheet with 3 tabs:
   - **Tab 1: Raw Data** - All SKUs with details (product_id, product_name, packing_unit_id, date, etc.)
   - **Tab 2: SKU Aggregation** - SKU level mapping percentage (mapped SKUs / total SKUs per product)
   - **Tab 3: Cat-Brand Aggregation** - Category-Brand level mapping percentage
4. Clears existing data before updating
5. Designed to run weekly

**Note:** Update the `workbook_name` variable in the date column cell with your Google Sheet name.


```python
# =============================================================================
# IMPORTS & SETUP
# =============================================================================
import pandas as pd
import os
import sys
from pathlib import Path

# Add parent directory to path to import common_functions
sys.path.insert(0, str(Path.cwd()))

from common_functions import snowflake_query, google_sheets, initialize_env

print("✓ Imports loaded successfully")
```

```python
# =============================================================================
# INITIALIZE ENVIRONMENT
# =============================================================================
print("Initializing environment...")
initialize_env()
print("✓ Environment initialized")
```

```python
# =============================================================================
# READ SQL QUERY
# =============================================================================
query_file = Path('queries') / 'cohort_700_sku_data.sql'

with open(query_file, 'r', encoding='utf-8') as f:
    query = f.read()

print(f"✓ Query loaded from {query_file}")
print(f"  Query length: {len(query)} characters")
```

```python
def convert_to_numeric(df):
    """Convert DataFrame columns to numeric where possible."""
    df.columns = df.columns.str.lower()
    for col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='ignore')
    return df
```

```python
# =============================================================================
# EXECUTE QUERY ON SNOWFLAKE
# =============================================================================
print("Executing query on Snowflake...")
print("(This may take a few moments...)")

df_raw = snowflake_query('Egypt', query)
df_raw = convert_to_numeric(df_raw)
print(f"✓ Query executed successfully")
print(f"✓ Retrieved {len(df_raw)} rows")
print(f"\nColumns: {list(df_raw.columns)}")

```

```python
# =============================================================================
# ADD DATE COLUMN WITH PRESERVATION LOGIC
# =============================================================================
from datetime import datetime

workbook_name = "Maxab X Savvy"  # Update this with the actual sheet name
today_date = datetime.now().strftime('%Y-%m-%d')

print("Reading existing sheet data to preserve dates...")
try:
    # Read existing data from the sheet
    existing_df = google_sheets(workbook_name, 'Raw Data', 'get')
    existing_df.columns = existing_df.columns.str.lower()
    
    if 'date' in existing_df.columns and 'product_id' in existing_df.columns:
        # Create a mapping of product_id to existing date
        existing_dates = existing_df[['product_id', 'date']].drop_duplicates()
        existing_dates['product_id'] = pd.to_numeric(existing_dates['product_id'], errors='coerce')
        
        # Merge with query results
        df_raw = df_raw.merge(existing_dates, on='product_id', how='left')
        
        # Fill missing dates with today's date
        df_raw['date'] = df_raw['date'].fillna(today_date)
        df_raw['date'] = df_raw['date'].replace('', today_date)
        
        print(f"✓ Preserved {df_raw['date'].ne(today_date).sum()} existing dates")
        print(f"✓ Added today's date ({today_date}) to {df_raw['date'].eq(today_date).sum()} new/missing entries")
    else:
        # No date column exists - add today's date to all
        df_raw['date'] = today_date
        print(f"✓ No existing date column found. Added today's date ({today_date}) to all {len(df_raw)} rows")
        
except Exception as e:
    # Sheet doesn't exist or is empty - add today's date to all
    df_raw['date'] = today_date
    print(f"✓ Could not read existing sheet. Added today's date ({today_date}) to all {len(df_raw)} rows")
    print(f"  (Reason: {str(e)})")

print(f"\nDate column summary:")
print(df_raw['date'].value_counts())
```

```python
sku_agg = df_raw.groupby('product_id').agg({
    'ben_soliman': 'max'
}).reset_index()

# Flatten column names
mapped = sku_agg[sku_agg['ben_soliman']==1]
mapped_count = len(mapped)
total = len(sku_agg)
percentage = round((mapped_count / total * 100)*2)/2

# Create summary DataFrame
summary_df = pd.DataFrame({
    'metric': ['Total SKUs', 'Mapped SKUs', 'Mapping Percentage'],
    'value': [total, mapped_count, f'{percentage}%']
})

print("Summary Statistics:")
display(summary_df)
print(f"\nMapping Percentage: {percentage}%")
```

```python
# =============================================================================
# CREATE CAT-BRAND LEVEL AGGREGATION
# =============================================================================
print("Creating Cat-Brand level aggregation...")

# Group by cat and brand
data = df_raw.groupby(['product_id','cat','brand']).agg({
    'ben_soliman': 'max'
}).reset_index()
cat_brand_agg = data.groupby(['cat', 'brand']).agg({
    'product_id': 'nunique',  # Count unique products
    'ben_soliman': 'sum'  # Sum of mapped SKUs
}).reset_index()

# Calculate total SKUs per cat-brand (count of all rows)
#total_skus = df_raw.groupby(['cat', 'brand']).size().reset_index(name='total_skus')

# Merge
#cat_brand_agg = cat_brand_agg.merge(total_skus, on=['cat', 'brand'])

# Rename columns
cat_brand_agg.columns = ['cat', 'brand', 'unique_products', 'mapped_skus']

# Calculate mapping percentage
cat_brand_agg['mapping_percentage'] = round((cat_brand_agg['mapped_skus'] / 
                                       cat_brand_agg['unique_products'] * 100)*2)/2

# Select and order columns
df_cat_brand_agg = cat_brand_agg[['cat', 'brand', 'unique_products', 
                                   'mapped_skus', 'mapping_percentage']]

# Sort by cat and brand
df_cat_brand_agg = df_cat_brand_agg.sort_values(['cat', 'brand'])

print(f"✓ Created aggregation for {len(df_cat_brand_agg)} cat-brand combinations")
print("\nFirst 5 rows of Cat-Brand aggregation:")
display(df_cat_brand_agg.head())
```

```python
# =============================================================================
# UPDATE GOOGLE SHEET
# =============================================================================
# Note: workbook_name is defined in the date column cell above

print(f"Updating Google Sheet: {workbook_name}")
print("=" * 70)

# Tab 1: Raw Data
print("\n1. Clearing and updating Tab 1: Raw Data...")
google_sheets(workbook_name, 'Raw Data', 'overwrite', df=df_raw)
print("   ✓ Tab 1 updated")

# Tab 2: SKU Aggregation
print("\n2. Clearing and updating Tab 2: SKU Aggregation...")
google_sheets(workbook_name, 'Overview', 'overwrite', df=summary_df)
print("   ✓ Tab 2 updated")

# Tab 3: Cat-Brand Aggregation
print("\n3. Clearing and updating Tab 3: Cat-Brand Aggregation...")
google_sheets(workbook_name, 'Cat-Brand Aggregation', 'overwrite', df=df_cat_brand_agg)
print("   ✓ Tab 3 updated")

print("\n" + "=" * 70)
print(f"✓ Successfully updated all tabs in {workbook_name}")
print("=" * 70)
```
